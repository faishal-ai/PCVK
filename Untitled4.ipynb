```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://media.istockphoto.com/id/1364630139/id/foto/anak-kucing-terisolasi-di-latar-belakang-hitam.jpg?s=612x612&w=0&k=20&c=4ISfyVXqfFpTe_cAq4P_3OqJ26RFOeNVg491u5U89Aw="]
# baca dan tampilkan image
# loop pada tiap url image
for url in urls:
  image = io.imread(url) #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB) #convert color to RGB
  final_frame = cv.hconcat((image, image_2)) #concatenate image
  cv2_imshow(final_frame) #show image
  print('\n')
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("relosuli image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2), y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis vertikal ditengah image
for x in range (tinggi):
  image_3[x, int(lebar/2)] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Garis X
tinggi, lebar, _ = image_3.shape

for x in range(lebar):
    y = int((tinggi / lebar) * x)
    if y < tinggi:
        image_3[y, x] = [255, 255, 255]

for x in range(lebar):
    y = int((-tinggi / lebar) * x + tinggi - 1)
    if 0 <= y < tinggi:
        image_3[y, x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Garis Pendek

start_x = 100
end_x   = 200

for y in range(start_x, end_x):
    image_3[int(tinggi/2), y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Tentukan koordinat kotak (x = kolom, y = baris)
x1, y1 = 60, 60
x2, y2 = 150, 150

for y in range(y1, y2):
    for x in range(x1, x2):
        image_3[y, x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
